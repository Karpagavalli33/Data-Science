Univariate Analysis

```python
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

## Bivariate Analysis-- Distribution Plot

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset)
plt.show()
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
```

```python
sb.pairplot(dataset,hue='specialisation',diag_kind="kde",kind="scatter",palette='husl')
plt.show()
```

#Categorical Plots 

```python
sb.stripplot(x='ssc_b',y='ssc_p',data=dataset)
```

```python
sb.stripplot(x='status',y='salary',data=dataset,jitter=True)
```

```python
sb.swarmplot(x='gender',y='ssc_p',data=dataset,color='red',size=7,hue='salary')
plt.show()
```

```python
sb.violinplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.violinplot(x="hsc_b",y="hsc_p",data=dataset,hue="gender")
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue="gender")
plt.show()
```

```python
sb.boxplot(x="ssc_b",y="ssc_p",data=dataset,hue='gender') ## it represents the percentile concept. 
plt.show()
```

##Bar Plot

```python
sb.barplot(x="ssc_b",y="ssc_p",data=dataset,hue='gender')
plt.show()
```

```python
sb.countplot(x="ssc_b",data=dataset,hue='gender')
plt.show()
```

```python
sb.pointplot(x="ssc_b",y="ssc_p",data=dataset,hue='gender')
plt.show()
```

```python
sb.boxplot(data=dataset,orient='h')
plt.show()
```
