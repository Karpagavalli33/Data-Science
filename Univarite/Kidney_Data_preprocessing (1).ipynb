```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("kidney_disease_test.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
def QuanQual(dataset):
    quan=[]
    qual=[]
    for ColumnName in dataset.columns:
        #print(ColumnName)
        if (dataset[ColumnName].dtypes=='O'):
            #print("qual")
            qual.append(ColumnName)
        else:
            #print("quan")
            quan.append(ColumnName)
    return quan,qual
```

```python
quan,qual=QuanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
dataset["rbc"].fillna(dataset["rbc"].mode(),inplace=True)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
df=pd.DataFrame(df,columns=quan)
```

```python
df
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(dataset[qual])
df1=imp.transform(dataset[qual])
df1=pd.DataFrame(df1,columns=qual)
```

```python
df1
```

```python
df1.isna().sum()
```

```python
merged_tables=[df,df1]
```

```python
result=pd.concat(merged_tables,axis=1)
```

```python
result
```

```python
result.isnull().sum()
```

```python
result.to_csv("Preprocessed.csv",index=False)
```
